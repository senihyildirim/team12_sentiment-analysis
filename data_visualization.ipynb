```python
#import libaries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud

```

```python
#import dataset
df = pd.read_csv('datasets/emotion_dataset.csv')
df.head(3)
```

```python
label_map={
    0:'sadness',
    1:'joy',
    2:'love',
    3:'anger',
    4:'fear',
    5:'surprise'
}
df['emotions'] = df['label'].map(label_map)
df.head(3)
```

```python
#plotting the wordcloud
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = None, 
                min_font_size = 10).generate(' '.join(df['text']))
plt.figure(figsize = (4, 4), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```

```python
#Emotion counts in the dataset
df['emotions'].value_counts().plot(kind='bar')
plt.title('Emotion Counts')
plt.xlabel('Emotions')
plt.ylabel('Counts')
plt.show()
```
